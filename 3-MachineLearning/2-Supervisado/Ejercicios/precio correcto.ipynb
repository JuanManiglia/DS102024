### Screen Resolution

```python
df["Screen Resolution"]
```

```python
df["Resolution"]=df["Screen Resolution"].str.extract(r"(\d+x\d+)")
```

```python
def categorize_resolution(pixels):
    if pixels <= 1049088:
        return "HD"
    elif pixels <= 2073600:
        return "Full HD"
    elif pixels <= 3686400:
        return "QHD"
    else:
        return "4K+"

df[["Width", "Height"]] = df["Resolution"].str.split("x", expand = True).astype(int)

df["Total_pixels"] = df["Width"] * df["Height"]

df["Resolution"] = df["Total_pixels"].apply(categorize_resolution)
```

```python
df
```

```python
df["Screen"] = df["Screen Resolution"].replace(r"(\d+x\d+)", "", regex=True)
```

```python
df["Screen"] = df["Screen"].replace(r"(Full HD|Quad HD|Quad HD|\+|/|4K Ultra HD)", "", regex=True)
```

```python
df["Touchscreen"] = df["Screen"].str.extract(r'(Touchscreen)')
df["Screen"] = df["Screen"].replace(r"(Touchscreen)", "", regex=True)
```

```python
df["Touchscreen"] = df["Touchscreen"].map({"Touchscreen": 1}).fillna(0)
```

```python
print(df["Touchscreen"].value_counts())
print(df["Screen"].value_counts())
```

```python
df["Screen"]=df["Screen"].replace(r" ", "", regex=True)
df["Screen"].value_counts()
```

```python
df["Screen"]=df["Screen"].replace(r"^\s*$", np.nan, regex=True)
df["Screen"].value_counts()
```

```python
df = df.drop(columns = ["Screen Resolution", "Width", "Height", "Total_pixels"])
```
