```python
# (\*\*raíz del tallo\*\*) ([a-zA-Z]+)|(\*\*superficie del tallo por encima del anillo\*\*) ([a-zA-Z]+)|(\*\*superficie del tallo por debajo del anillo\*\*) ([a-zA-Z]+)|(\*\*color del tallo por encima del anillo\*\*) ([a-zA-Z]+)|(\*\*color del tallo por debajo del anillo\*\*) ([a-zA-Z]+)|(\*\*tipo de velo\*\*) ([a-zA-Z]+)|(\*\*color de velo\*\*) ([a-zA-Z]+)|(\*\*anillo(s)\*\*) (\d+)|(\*\*de tipo\*\*) ([a-zA-Z]+)|(\*\*color de la impresión de esporas\*\*) ([a-zA-Z]+)|(\*\*población\*\*) ([a-zA-Z]+)|(\*\*hábitat es\*\*) ([a-zA-Z]+)

df["observacion"] = df["descripcion"].str.extract(r"(\d+)")
df["clase"] = df["descripcion"].str.extract(r"\*\*clase\*\* ([a-zA-Z]+)")
df["forma de sombrero"] = df["descripcion"].str.extract(r"\*\*forma de sombrero\*\* ([a-zA-Z]+)")
df["superficie"] = df["descripcion"].str.extract(r"\*\*superficie\*\* ([a-zA-Z]+)")
df["color"] = df["descripcion"].str.extract(r"\*\*y un color\*\* ([a-zA-Z]+)")
df['moretones'] = df['descripcion'].apply(lambda x: 'Sí' if 'Presenta' in x else ('No' if 'No presenta' in x else None))
df["olor"] = df["descripcion"].str.extract(r"\*\*olor\*\* ([a-zA-Z]+)")
df["Tipo de branquias"] = df["descripcion"].str.extract(r"\*\*Las branquias son de tipo\*\* ([a-zA-Z]+)")
df["separación branquias"] = df["descripcion"].str.extract(r"\*\*separación\*\* ([a-zA-Z]+)")
df["tamaño branquias"] = df["descripcion"].str.extract(r"\*\*y un tamaño\*\* ([a-zA-Z]+)")
df["color branquias"] = df["descripcion"].str.extract(r"\*\*color de las branquias \*\* ([a-zA-Z]+)")
df["forma del tallo"] = df["descripcion"].str.extract(r"\*\*forma del tallo\*\* ([a-zA-Z]+)")
df["raíz del tallo"] = df["descripcion"].str.extract(r"\*\*raíz del tallo\*\* ([a-zA-Z]+)")
df.head()
```
