5. Crea todas las funciones de una calculadora (suma, resta, multiplicaci칩n, divisi칩n)

```python
def suma (*numeros:float):
    '''
    Le vamos a dar una variable con uniciador 0 
    mientras va recorriendo cada numero de la cadena que se le se (numeros)
    la vamos sumando y guardando en la variable suma
    '''
    suma=0
    for var in numeros:
        suma=suma + var
    return suma
```

```python
def resta(*numeros:float):
    '''
    iniciamos la variable 'resta' con el primer elemento que se le pasa,
    para que no se reste la misma variable la de la posicuion 0 , verificamos 
    que empezamos a restar a partir de la posicion 1 'if var!=numeros[0]:'
    y lo vamos guardando en la variable 'resta'
    '''
    resta=numeros[0]

    for var in numeros:
        if var!=numeros[0]:
            resta=resta-var
        
    return resta
```

```python
def multiplicacion(*numeros:float)->float:
    '''
    iniciamos la variable 'multi' con el primer elemento que se le pasa,
    para que no se multiplique la misma variable la de la posicuion 0 , verificamos 
    que empezamos a multiplicar a partir de la posicion 1 'if var!=numeros[0]:'
    y lo vamos guardando en la variable 'multi'
    '''
    multi=numeros[0]
    for var in numeros:
        if var!=numeros[0]:
            multi=multi*var
    return multi
```

```python
def division(*numeros:float)->float:
    '''
    iniciamos la variable 'divisor' con el primer elemento que se le pasa,
    para que no se dividen la misma variable la de la posicuion 0 , verificamos 
    que empezamos a dividir a partir de la posicion 1 'if var!=numeros[0]:'
    y lo vamos guardando en la variable 'divisor'
    '''
    divisor=numeros[0]
    for var in numeros:
        if var!=numeros[0]:
            divisor=divisor/var
    return divisor
```

```python
print(suma(2,3,10,20))
print(resta(2,8,3))
print(multiplicacion(2,5,10))
print(division(100,2,5))
```
